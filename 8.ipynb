```python
# ==========================================
# Global Dependencies Import

import os                        
import cv2                       
import numpy as np               
import matplotlib.pyplot as plt  

print("(Environment setup successful!) All libraries loaded.")
```

```python

#  Data Loading Pipeline


image_folder = "Image" 
dataset = {}           

if not os.path.exists(image_folder):
    print(f"Error: Directory '{image_folder}' not found!") 
else:
    for file_name in os.listdir(image_folder):
        if file_name.lower().endswith(('.jpg', '.jpeg')):
            
            file_path = os.path.join(image_folder, file_name)
            img_bgr = cv2.imread(file_path) 
            
            if img_bgr is not None:
                # Color Space Conversion
                img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
                dataset[file_name] = img_rgb 
                print(f"✅ Successfully loaded: {file_name} | Shape: {img_rgb.shape}")
            else:
                print(f"❌ Warning: Cannot read {file_name}.")

    print("=" * 50)
    print(f"Data Pipeline Ready! Total images loaded: {len(dataset)}")
```
